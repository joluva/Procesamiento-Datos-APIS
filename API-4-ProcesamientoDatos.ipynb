4 - Ejecutar el siguiente ejercicio, compartir una captura de pantalla,
cambiando el título y modificando cantidades de frutas.

```python
# cambio las cantidades en las frutas, el titulo y agrego linea de codigo para alinear las etiquetas
data = {'manzanas': 15, 'naranjas': 25, 'limones': 8, 'peras': 16}
Nombres= list(data.keys())
Valores= list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(Nombres, Valores)
axs[0].tick_params(axis='x', labelrotation=45)
axs[0].grid(True, linestyle='--', color='gray', linewidth=0.5)


axs[1].scatter(Nombres, Valores)
axs[1].tick_params(axis='x', labelrotation=45)
axs[1].grid(True, linestyle='--', color='gray', linewidth=0.5)


axs[2].plot(Nombres, Valores)
axs[2].tick_params(axis='x', labelrotation=45)
axs[2].grid(True, linestyle='--', color='gray', linewidth=0.5)


fig.suptitle('Frutas - Cantidades')
plt.show()
```
